```python
import matplotlib.pyplot as plt
import numpy as np
# plt.style.use("ggplot")  
```

```python
seq=15128410
a=1
b=seq/6106174
c=seq/3086648

num_process=[1,2,3]
ideal=[1,2,3]
plt.plot(num_process, [a,b,c], 's-', label="speedup")
plt.plot(num_process, ideal, 's-', label="ideal")
plt.xlabel("# worker nodes")
plt.ylabel("speedup factor")
plt.xticks(num_process,num_process)
plt.legend()
plt.show()
```

```python
computing_time=[15128410, 6106174, 3086648]
num_process=[1,2,3]
plt.bar(num_process, computing_time)
plt.xlabel("# worker nodes")
plt.ylabel("runtime (microseconds)")
plt.legend()
plt.show()
```
