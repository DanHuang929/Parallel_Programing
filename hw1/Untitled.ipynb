```python
import matplotlib.pyplot as plt
# plt.style.use("ggplot")  
```

```python
seq=29.424
a=1
b=seq/16.5464
c=seq/13.5464
d=seq/10.507
e=seq/10.170
num_process=[1,2,4,8,12]
ideal=[1,2,4,8,12]
plt.plot(num_process, [a,b,c,d,e], 's-', label="speedup")
plt.plot(num_process, ideal, 's-', label="ideal")
plt.xlabel("# processes")
plt.ylabel("speedup factor")
plt.xticks(num_process,num_process)
plt.legend()
plt.show()
```
