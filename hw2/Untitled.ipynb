```python
import matplotlib.pyplot as plt
# plt.style.use("ggplot")  
```

```python
seq=72132763
a=1
b=seq/36186425
c=seq/18121251
d=seq/9502020
e=seq/6123898
num_process=[1,2,4,8,12]
ideal=[1,2,4,8,12]
plt.plot(num_process, [a,b,c,d,e], 's-', label="speedup")
plt.plot(num_process, ideal, 's-', label="ideal")
plt.xlabel("# threads")
plt.ylabel("speedup factor")
plt.xticks(num_process,num_process)
plt.legend()
plt.show()
```
