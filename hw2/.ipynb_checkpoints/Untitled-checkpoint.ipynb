```python
import matplotlib.pyplot as plt
# plt.style.use("ggplot")  
```

```python
seq=72132763
a=1
b=seq/36186425
c=seq/18121251
d=seq/9502020
e=seq/6123898
num_process=[1,2,4,8,12]
ideal=[1,2,4,8,12]
plt.plot(num_process, [a,b,c,d,e], 's-', label="speedup")
plt.plot(num_process, ideal, 's-', label="ideal")
plt.xlabel("# threads")
plt.ylabel("speedup factor")
plt.xticks(num_process,num_process)
plt.legend()
plt.show()
```

```python
IO_time=[45727, 44405, 75846, 78690, 106799]
computing_time=[72132763, 36186425, 18121251, 9502020, 6123898]
# comm_time=[0, 0.760, 2.344, 3.340, 3.373]
num_process=['1','2','4','8','12']
plt.bar(num_process, IO_time, label="IO time")
plt.bar(num_process, computing_time, bottom=IO_time, label="computing time")
# plt.bar(num_process, comm_time, bottom=computing_time, label="communication time")
plt.xlabel("# processes")
plt.ylabel("runtime (microseconds)")
plt.legend()
plt.show()
```
