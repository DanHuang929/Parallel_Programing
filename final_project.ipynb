```python
import random
import matplotlib.pyplot as plt
```

```python
alphbat = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
print(len(alphbat))
```

```python
LCS = "iamdanandiloveparallelprogramming"
print(len(LCS))
```

```python
bound=300

A=""
index=0
while(True):
    if(index==len(LCS)):
        break
    put_LCS = random.randint(0, bound)
    if(put_LCS==1):
        A+=LCS[index]
        index+=1
    else:
        alph = random.randint(0, 25)
        A+=alphbat[alph]
print(len(A))
print(A)

B=""
index=0
while(True):
    if(index==len(LCS)):
        break
    put_LCS = random.randint(0, bound)
    if(put_LCS==1):
        B+=LCS[index]
        index+=1
    else:
        alph = random.randint(0, 25)
        B+=alphbat[alph]
print(len(B))
print(B)
```

```python
ans = lcs(A, B)
print ("Length of LCS is ", lcs(A, B))
```

```python
id+=1

path = 'input'+str(id)+'.txt'
f = open(path, 'w')
f.write(A)
f.write("\n")
f.write(B)
f.close()

path = 'output'+str(id)+'.txt'
f = open(path, 'w')
f.write(str(ans))
f.close()
```

```python
def lcs(X, Y):
    # find the length of the strings
    m = len(X)
    n = len(Y)
 
    # declaring the array for storing the dp values
    L = [[None]*(n + 1) for i in range(m + 1)]
 
    """Following steps build L[m + 1][n + 1] in bottom up fashion
    Note: L[i][j] contains length of LCS of X[0..i-1]
    and Y[0..j-1]"""
    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0 or j == 0 :
                L[i][j] = 0
            elif X[i-1] == Y[j-1]:
                L[i][j] = L[i-1][j-1]+1
            else:
                L[i][j] = max(L[i-1][j], L[i][j-1])
 
    # L[m][n] contains the length of LCS of X[0..n-1] & Y[0..m-1]
    return L[m][n]

 
X = "AGGTAB"
Y = "GXTXAYB"
print ("Length of LCS is ", lcs(X, Y))
```
